```python
import gzip
import json
import re
import os
import sys
import numpy as np
import pandas as pd
```

```python
DIR = './'
```

```python
def load_data(file_name, head = 1000):
    count = 0
    data = []
    with gzip.open(file_name) as fin:
        for l in fin:
            d = json.loads(l)
            count += 1
            data.append(d)
            
            # break if reaches the 100th line
            if (head is not None) and (count > head):
                break
    return data
```

```python
children = load_data(os.path.join(DIR, 'goodreads_books_children.json.gz'))
comics = load_data(os.path.join(DIR, 'goodreads_books_comics_graphic.json.gz'))
fantasy = load_data(os.path.join(DIR, 'goodreads_books_fantasy_paranormal.json.gz'))
history = load_data(os.path.join(DIR, 'goodreads_books_history_biography.json.gz'))
mystery = load_data(os.path.join(DIR, 'goodreads_books_mystery_thriller_crime.json.gz'))
poetry = load_data(os.path.join(DIR, 'goodreads_books_poetry.json.gz'))
romance = load_data(os.path.join(DIR, 'goodreads_books_romance.json.gz'))
yadult = load_data(os.path.join(DIR, 'goodreads_books_young_adult.json.gz'))
```
